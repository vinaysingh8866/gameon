```python

import json
# with open(fileN+tag['src'], "rb", buffering=0) as file_handler:
#             tx = Transaction(wallet, file_handler=file_handler, file_path=fileN+tag['src'])
#             tx.add_tag('Content-Type','application/javascript') 
#             tx.sign()
#             uploader = get_uploader(tx, file_handler)
#             while not uploader.is_complete:
#                 uploader.upload_chunk()

#                 print("{}% complete, {}/{}".format(
#                     uploader.pct_complete, uploader.uploaded_chunks, uploader.total_chunks
#                 ))
#             tagsDict[tx.id] = tag['src']
#             print(tag, tx.id)

def uploadNFTImage(fileN):
    with open(fileN+tag['src'], "rb", buffering=0) as file_handler:
        tx = Transaction(wallet, file_handler=file_handler, file_path=fileN+tag['src'])
        tx.add_tag('Content-Type','image/png')
        tx.sign()
        uploader = get_uploader(tx, file_handler)
        while not uploader.is_complete:
            uploader.upload_chunk()
            print("{}% complete, {}/{}".format(
                uploader.pct_complete, uploader.uploaded_chunks, uploader.total_chunks
            ))
        return tx.id


def uploadNFTJSON(fileN):
    with open(fileN+tag['src'], "rb", buffering=0) as file_handler:
        tx = Transaction(wallet, file_handler=file_handler, file_path=fileN+tag['src'])
        tx.add_tag('Content-Type','application/json')
        tx.sign()
        uploader = get_uploader(tx, file_handler)
        while not uploader.is_complete:
            uploader.upload_chunk()
            print("{}% complete, {}/{}".format(
                uploader.pct_complete, uploader.uploaded_chunks, uploader.total_chunks
            ))
        return tx.id

def changeURLToArweave(url,file):
    f = open(file, "r")
    contents = f.read()
    f.close()
    #load json
    data = json.loads(contents)
    #change url
    data['image'] = url
    #save json
    with open(file, 'w') as outfile:
        json.dump(data, outfile)

images = glob.glob("nfts/images/*")
for image in images:
    print(image)
    txId = uploadNFTImage(image)
    print(txId)
    ar = "https://arweave.net/" + txId
    changeURLToArweave(ar,image)
    uploadNFTJSON("nfts/json/" + image.split("/")[-1].split(".")[0] + ".json")

```
